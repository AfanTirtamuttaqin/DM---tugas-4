```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error,root_mean_squared_error
```

```python
data = pd.read_excel('Tanaman Padi.xlsx', usecols=['Tahun', 'Produksi Tanaman Padi'])
```

```python
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data["Lag_produksi"] = data["Produksi Tanaman Padi"].shift(1)
```

```python
data.head()
```

```python
data["Lag_produksi"].fillna(data["Lag_produksi"].mode()[0], inplace=True)
```

```python
data.head()
```

```python
x = data['Lag_produksi'].values.reshape(-1,1)
y = data['Produksi Tanaman Padi'].values.reshape(-1,1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
lin_reg = LinearRegression()
```

```python
lin_reg.fit(x_train, y_train)
```

```python
LinearRegression()
```

```python
acc = lin_reg.score(x_test, y_test) *100
acc
```

```python
mse = mean_squared_error(y_test, x_test)
rms = root_mean_squared_error(y_test, x_test)
mape = mean_absolute_percentage_error(y_test, x_test)
```

```python
mse,rms,mape
```
